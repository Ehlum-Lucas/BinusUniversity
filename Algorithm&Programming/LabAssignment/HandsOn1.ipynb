```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x, y, color='#ff6347', linestyle='--', linewidth=2, marker='o', label='Prime Numbers')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot of Prime Numbers')
plt.annotate('Max Value', xy=(5, 11), xytext=(4, 10),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.grid(True)
plt.legend()
plt.savefig('line_plot.png')
plt.show()
```

```python
x = [10, 20, 30, 40, 50]
y = [15, 25, 35, 45, 55]
plt.scatter(x, y, color='#1e90ff', s=50, marker='^', label='Data Points')
plt.title('Scatter Plot with Triangle Markers')
plt.annotate('Point (40, 45)', xy=(40, 45), xytext=(30, 50),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True)
plt.legend()
plt.savefig('scatter_plot.png')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 10, 15, 20, 25]
plt.bar(categories, values, color='#32cd32', width=0.5, label='Bar Chart')
plt.title('Bar Chart with Custom Width')
for i, value in enumerate(values):
    plt.text(i, value + 0.5, str(value), ha='center')
plt.grid(True, axis='x')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.legend()
plt.savefig('bar_chart.png')
plt.show()
```
